```python
First = int(input("starting number"))
Last = int(input("Endind number"))
even = []
odd = []

for i in range(First, Last+1):
    if i%2==0:
        even.append(i)
print("Number of even numbers=",len(even))
for i in range(First, Last+1):
    if i%2!=0:
        odd.append(i)
print("Number of odd numbers=",len(odd))        

```
